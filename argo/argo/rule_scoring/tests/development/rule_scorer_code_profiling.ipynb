```python
import pandas as pd
import numpy as np
from rule_scoring.rule_scorer import RuleScorer
import cProfile, pstats
```

```python
om = pd.read_pickle('~/Downloads/omnyex_processed_for_profiling.pkl')
X = om.drop('chargeback', axis=1)
y = om['chargeback']
```

```python
X = X.astype(float)
```

```python
X_rules = (X >= 1).astype(int)
```

```python
from rule_scoring.rule_scoring_methods import PerformanceScorer, LogRegScorer, RandomForestScorer
from rule_scoring.rule_score_scalers import ConstantScaler, MinMaxScaler
from rule_optimisation.optimisation_functions import Precision
```

```python
X_rules.shape
```

# PerformanceScorer with ConstantScaler

## Old

```python
precision = Precision()
```

```python
# rs = RuleScorer(scoring_class=PerformanceScorer(performance_func=precision.fit), scaling_class=ConstantScaler(limit=-100))
# rs.fit(X_rules, y)
```

```python
filename = 'PerformanceScorer_ConstantScaler_old.dat'
rs = RuleScorer(scoring_class=PerformanceScorer(performance_func=precision.fit), scaling_class=ConstantScaler(limit=-100))
cProfile.run('rs.fit(X_rules, y)', sort='cumtime', filename=f'{filename}.dat')
```

```python
p = pstats.Stats(f'{filename}.dat')
p.sort_stats('cumtime').print_stats()
```

~.7sec runtime - pretty quick already
